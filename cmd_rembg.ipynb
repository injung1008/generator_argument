```python
import os

import glob
image_path = "./apple/"
image_path_list = []
for filename in glob.glob(f'{image_path}*.png'): #assuming gif
    image_path_list.append(filename)

for img_path in image_path_list:
    cmd = f'rembg i -m isnet-general-use "{img_path}" "{img_path.replace("apple","output_apple")}"'
    os.system(cmd)
```
